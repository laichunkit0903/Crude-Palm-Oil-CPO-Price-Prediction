# Features Engineering



<b>Determnine corrrelation between features and target , elimate those features which are not significantly correlate with the target</b>

```python
#Lets put columns other than PAL2MALY Index_PX_LAST and date as features
features = month_df[month_df.columns.difference(['Date','PAL2MALY Index_(Crude Palm Oil fob Spot)'])]
#target = month_df[['PAL2MALY Index_(Crude Palm Oil fob Spot)']]
```

```python
features.head()
```

```python
#Find out corelation between selected features and targets using pearson correlation and determine signifincace level using p-Value
result =[]
for i in features.columns:
    result.append((i,scipy.stats.pearsonr(np.array(features[i]), np.array(month_df['PAL2MALY Index_(Crude Palm Oil fob Spot)']))))
    
result
```

Based on the above results, beisdes <b>QS1 Comdty_(ICE Gas Oil Future)</b> and <b>PASTTOTL Index_(Total Palm Oil Inventory Data)</b>, others columns are significantly correaltes with target columns as other columns p-value are < <b>0.05</b>.

```python
#Using heatmap to inspect correlation between features ans targets

plt.figure(figsize=(12,10))
corr = month_df[month_df.columns.difference(['Date'])].corr()
sns.heatmap(corr, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
#make the result absolute
cor_target = abs(corr["PAL2MALY Index_(Crude Palm Oil fob Spot)"])


cor_target
```

```python
#Inspect Features that have correlation > 0.5 (include <-0.5 as abcolute value)
relevant_features = cor_target[(cor_target>0.5)]
relevant_features
```

BO1 Comdty_(CBOT Soybean Oil Future) , CO1 Comdty_(ICE Brent Crude Oil Future)  and QS1 Comdty_(ICE Gas Oil Future) are highly significant features that could help in predict CPO spot price
