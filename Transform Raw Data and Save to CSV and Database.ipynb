```python
#Import required package
import pandas as pd
import numpy as np

#Read the CSV file without first row of title
month_df = pd.read_csv('data_set/Assignment-monthlyexcerpt.csv',skiprows=1,low_memory=False)
daily_df = pd.read_csv('data_set/Assignment-dailyexcerpt.csv',skiprows=1,low_memory=False)

#Read the first row of the CSv file
month_df_first_header = pd.read_csv('data_set/Assignment-monthlyexcerpt.csv', nrows=0,low_memory=False)
daily_df_first_header = pd.read_csv('data_set/Assignment-dailyexcerpt.csv',nrows=0,low_memory=False)
```

```python
#Inspect first row of the header
month_df_first_header.columns
```

```python
#Remove columns name with 'Unnamed'
month_df_first_header = [a for a in month_df_first_header.columns if 'Unnamed' not in a ]
month_df_first_header
```

```python
#Combine every two name to become meaningful column
month_df_useful_heder = []
for i,k in zip(month_df_first_header[0::2], month_df_first_header[1::2]):
    month_df_useful_heder.append(i+ "_(" + k + ")")
    
month_df_useful_heder
```

```python
#inspect monthly data frame
month_df.head()
```

Just use one date column and combine all the columns together since this date columns are the same

```python
##Use the meaningful ttitle to fit in monthly data frame column
def combine_title(df):
    date_col = df.Date
    col = df.columns
    
    #Drop all date columns
    for x in range(0, len(col)):
        if('Date' in col[x] or 'PX_LAST' not in col[x]):
            df = df.drop([col[x]], axis =1)
            
    #Replace remaining columns with meainingful columns created above
    for i in range(0, len(month_df_useful_heder)):
        df = df.rename(columns ={df.columns[i]:month_df_useful_heder[i]})
            
    df.insert(loc=0, column='Date', value=date_col)
    
    return df
        
```

```python
##Implement the fucntion
clean_month_df = combine_title(month_df)

clean_month_df.head()
```

```python
#write the result data to csv
clean_month_df.to_csv("data_set/combine_monthlyexcerpt.csv", index = False)
```

```python
#Take meeaningful title
daily_df_first_header = daily_df_first_header.columns[::3]
daily_df_first_header
```

```python
#Split to different data frame and rename them 
df_list = []
for i,k in zip(daily_df.columns[0::2], daily_df.columns[1::2]):
    df_list.append(daily_df[[i,k]])
    
for j in range(0, len(daily_df_first_header)):
    df_list[j] = df_list[j].rename(columns = {df_list[j].columns[0] : 'Date'})
    df_list[j] = df_list[j].rename(columns = {df_list[j].columns[1] : daily_df_first_header[j] + '_PX_LAST'})
    df_list[j].to_csv("data_set/" + daily_df_first_header[j] +".csv", index =False)
```
